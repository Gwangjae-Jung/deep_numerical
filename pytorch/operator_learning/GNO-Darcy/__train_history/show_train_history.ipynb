```python
import  glob
import  matplotlib.pyplot   as  plt
import  numpy     as  np
import  pickle


LINESTYLE = ['r-', 'r--', 'g-', 'g--']
```

```python
HISTORY_FILE = glob.glob("*.pickle")
HISTORY_NAME = tuple(file.split('.')[0] for file in HISTORY_FILE)
for name in HISTORY_NAME:
    print("*", name)
```

```python
HISTORY_DICT = {}
for name, file in zip(HISTORY_NAME, HISTORY_FILE):
    with open(file, "rb") as f:
        HISTORY_DICT[name] = pickle.load(f)

NUM_EPOCHS = len(HISTORY_DICT[HISTORY_NAME[0]]['train_loss'])
EPOCH_ARRAY = np.arange(NUM_EPOCHS) + 1
```

Training time

```python
for name in HISTORY_NAME:
    print(f"* {name:30s}: {HISTORY_DICT[name]['train_time']:.2f}s")
```

Train loss

```python
plt.figure(figsize = (15, 6))
plt.title("Train loss (MSE)")
for cnt, name in enumerate(HISTORY_NAME):
    plt.plot(EPOCH_ARRAY, HISTORY_DICT[name]['train_loss'], LINESTYLE[cnt], label = name)
plt.legend()
plt.yscale('log')
plt.show()
```

Train error

```python
plt.figure(figsize = (15, 6))
plt.title("Train error (relative $L^2$ error)")
for cnt, name in enumerate(HISTORY_NAME):
    plt.plot(EPOCH_ARRAY, HISTORY_DICT[name]['train_error'], LINESTYLE[cnt], label = name)
plt.legend()
plt.yscale('log')
plt.show()
```

Validation loss

```python
plt.figure(figsize = (15, 6))
plt.title("Validation loss (MSE)")
for cnt, name in enumerate(HISTORY_NAME):
    plt.plot(EPOCH_ARRAY, HISTORY_DICT[name]['val_loss'], LINESTYLE[cnt], label = name)
plt.legend()
plt.yscale('log')
plt.show()
```

Validation error

```python
plt.figure(figsize = (15, 6))
plt.title("Validation error (relative $L^2$ error)")
for cnt, name in enumerate(HISTORY_NAME):
    plt.plot(EPOCH_ARRAY, HISTORY_DICT[name]['val_error'], LINESTYLE[cnt], label = name)
plt.legend()
plt.yscale('log')
plt.show()
```
